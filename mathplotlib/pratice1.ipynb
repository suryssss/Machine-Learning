```python
import matplotlib.pyplot as plt
import numpy as np

x=[2021,2022,2023,2024,2025]
y=[10,30,5,40,60]

plt.plot(x,y)
plt.show()
```

```python
#markers

x=np.array([2021,2022,2023,2024,2025])
y1=np.array([10,35,20,40,60])
y2=np.array([25,40,60,20,5])

line_style=dict(marker=".",ms=20,mfc='#8ccf11',mec="white",ls="dashdot",lw=3)

plt.plot(x,y1,marker=".",ms=20,mfc='#8ccf11',mec="white",ls="dashdot",lw=3,color="red") #ms=markersize mfc=markerfacecolor mec=markeredgecolor ls=linestyle #lw=linewidth

plt.plot(x,y2,**line_style,color="yellow") # **unpacking of dictinary
plt.show() 
```

```python
#labels

x=np.array([2000,2001,2002,2003,2004])
y1=np.array([10,30,40,20,15])
y2=np.array([50,10,25,45,60])

l_style=dict(marker="*",ms=10,mfc="green",mec="white",ls="solid",lw=3)

plt.title("Class Size",fontsize=20,family='Arial',fontweight='bold',color="red")

plt.xlabel("Year",fontsize=20,family='Arial',fontweight='bold',color="blue")
plt.ylabel("No of students",fontsize=20,family='Arial',fontweight='bold',color="blue")

plt.tick_params(axis="both",colors="green",size)

plt.plot(x,y1,**l_style)
plt.plot(x,y2,**l_style)

plt.xticks(x) # to only show whats in the array

plt.show()
```

```python
#grid lines
#grid()=helps plots make easier to read by adding reference lines

x=[1,2,3,4,5]
y=[5,10,15,20,25]

plt.grid(axis='x',linewidth=2,color="lightgrey",linestyle="dashdot")

plt.plot(x,y)

plt.show()
```

```python
#barchats

#represents data in a bar

categories=np.array(["Grapes","Watermelon","Apple","Guava","Banana"])
values=np.array([200,120,300,156,345])

plt.title("Favourite Fruits")
plt.xlabel("Type of Fruit")
plt.ylabel("No of people like the fruit")

plt.bar(categories,values,color="lightblue")  #vertical bar chart
# plt.barh(categories,values,color="lightblue") #horizonal bar chart 


plt.show
```

```python
#Pie charts

#circular representation of data

categories=np.array(["1st year","2nd year",'3rd year',"4rth year"])
values=np.array([300,250,275,225])
colors=["red","blue","green","yellow"]

plt.title("College Students")


plt.pie(values,labels=categories,autopct="%1.1f%%",colors=colors,explode=[0,0,0.1,0],shadow=True,startangle=90)

plt.show()
```

```python
#scatterplots
# shows the relationship btw two variables 

x=np.array([0,1,1,2,3,4,5,6,7,7,8,9,9,9])
y=np.array([55,60,65,62,68,70,75,78,82,85,87,90,92,93])

plt.title("Test scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grades")

plt.scatter(x,y,color="red",alpha=0.75,s=200)

plt.show()
```

```python
#histograms
# visual representation of distrubution of quantitative data

scores=np.random.normal(loc=80,scale=10,size=100)
scores=np.clip(scores,0,100)

plt.hist(scores,bins=20,color="red",edgecolor="black")
plt.title("Exam Scores")
plt.xlabel("scores")
plt.ylabel("No of Students")

plt.show()
```

```python
#subplots
figure,axes=plt.subplots(2,3)
x=np.array([1,2,3,4,5])

axes[0,0].plot(x,x*2,color="red")
axes[0,0].set_title("x*2")
axes[0,1].plot(x,x*x,color="green")
axes[0,1].set_title("x*x")
axes[0,2].plot(x,x**0.5,color="yellow")
axes[0,2].set_title("x**0.5")

axes[1,0].bar(x,x**4)
axes[1,1].pie(x**4)

plt.tight_layout()
plt.show()
```

```python
#pandas+matplotlib

import pandas as pd

df=pd.read_csv("data.csv")


type_count=df["Type1"].value_counts(ascending=True)

plt.title("No of pokemons based on type")
plt.xlabel("No of pokemons")
plt.ylabel("Pokemon Type")
plt.barh(type_count.index,type_count.values,color="red",edgecolor="black")

plt.tight_layout()
plt.show()
```
