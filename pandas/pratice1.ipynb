```python
#series - a pandas 1-dimensional labelled array that can hold any data type think of it like a single 
#column of data in a spreadsheet (1-dimensional array with labels)

data=[100,101,102,400]

series=pd.Series(data,index=["a","b","c","d"])
print(series)

print(series.loc["c"])

series.loc["a"]=200
series.loc["e"]=300
print(series)
print(series.iloc[3])

print(series[series>200])

calories={"Day1":420,"Day2":380,"Day3":390}

series=pd.Series(calories)

series.loc["Day3"]+=450
print(series)
```

```python
#DataFrames- A tabular data structure with rows and columns (2-dimensional) similar to excel spreadsheet

data={"day":["Day1","Day2","Day3"],"calories":[420,380,390],"duration":[50,40,45]}
df=pd.DataFrame(data,index=["a","b","c"])

#add new column
df.loc["d"]=["Day4",450,60]
print(df)
print(df.loc["a"])
print(df.iloc[1])

#add new row
new_row=pd.DataFrame([{"day":"Day5","calories":500,"duration":70},
                      {"day":"Day6","calories":550,"duration":75}],index=["e","f"])
df=pd.concat([df,new_row])
print(df)
```

```python
#filtering - keeping the rows that matches a condition 

import pandas as pd
df=pd.read_csv("data.csv")

tall_pokemon=df[df["Height"]>2.0]

water_pokemon=df[(df["Type1"]=="Water") | (df["Type2"]=="Water")]
print(tall_pokemon)
print(water_pokemon)
```

```python
#aggregation - reduces a set of values to a single summary value used to summerize and analayze data often used with the 
#groupby() function 
import pandas as pd
df=pd.read_csv("data.csv")

#whole data frame
print(df.mean(numeric_only=True))
print(df.min(numeric_only=True))
print(df.sum(numeric_only=True))
print(df.max(numeric_only=True))

#single column 
print(df["Height"].mean())
print(df["Weight"].max())

#groupby - allows you to group data based on one or more columns and then apply aggregation functions to each group

group=df.groupby("Type1")
print(group["Height"].mean())
print(group["Weight"].max())
```

```python
#data cleaning - the process of identifying and correcting errors or inconsistencies in a dataset to improve its quality and reliability

import pandas as pd
df=pd.read_csv("data.csv")

df=df.drop(columns=["No"])

#handle missing values
# df=df.dropna(subset=["Type2"])
df=df.fillna({"Type2":"None"})

#fix inconsistent values
df["Type1"]=df["Type1"].replace({"Grass":"GRASS"})

#standardize units
df["Name"]=df["Name"].str.upper()
df["Height"]=df["Height"]*0.1

#fix data types

df["Lengandary"]=df["Legendary"].astype(bool)

#remove duplicates
df=df.drop_duplicates()
print(df)
```
